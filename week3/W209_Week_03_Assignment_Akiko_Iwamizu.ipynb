```
import altair as alt
```

```
alt.__version__
```

```
from vega_datasets import data

weather = data.seattle_weather()
weather.head()
```

```
len(weather)
```

```
# Hover over the points for the tooltips!
alt.Chart(weather).mark_circle(size=50).encode(
    x=alt.X('wind:Q', title="Wind"),
    y=alt.Y('precipitation:Q', title="Precipitation"),
    color=alt.Color('weather', 
                    legend=alt.Legend(title="Weather Category",
                                      orient="right",
                                      labelFontSize=14,
                                      titleFontSize=14),
                    scale=alt.Scale(scheme='dark2')
                    ),
    tooltip=['weather', 'precipitation', 'wind']
).configure_axis(labelFontSize=14,
                 titleFontSize=14
                 ).properties(title='Seattle Weather: Precipitation x Wind'
                 ).interactive()
```
